```python
allegations_officers = spark.sql("SELECT allegation_id,officer_id from officer_allegation_csv")
allegations = spark.sql("select ID, incident_date as date, beat_id as beat from allegations")
rank_salary = spark.sql("select id, rank, salary from salary_csv")
tempA = allegations_officers.join(rank_salary,allegations_officers['officer_id'] == rank_salary['id']).drop('id')
df = tempA.join(allegations,tempA['allegation_id'] == allegations['ID']).drop('ID')
rank_salary = spark.sql("select rank, salary from salary_csv")
rank_salary = rank_salary.groupby("rank").agg({'salary':'mean'})
```

```python
from pyspark.sql.functions import datediff, to_date, lit, round
officer_complaints = df.groupby("officer_id").count()
maxDates = df.groupby("officer_id").agg({"date":"max"})
minDates = df.groupby("officer_id").agg({"date": "min"})
dateRanges = minDates.join(maxDates,"officer_id").join(officer_complaints,"officer_id").withColumn("Date Range", datediff("max(date)", "min(date)"))
dateRanges2 = dateRanges.withColumn("Ratio", round(dateRanges["Date Range"] / dateRanges["count"], 2))
display(dateRanges2)
```
